# **Cleaning Data:**
1. URLs
2. hashtags
3. mentions
4. special letters
5. punctuations

```python
import re

def cleanResume(txt):
  # for urls
  cleanTxt = re.sub('http\S+\s', '', txt)
  # for hashtags
  cleanTxt = re.sub('#\S+', '', cleanTxt)
  # for mentions
  cleanTxt = re.sub('@\S+', '', cleanTxt)
  # for special words
  cleanTxt = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanTxt)

  cleanTxt = re.sub(r'[^\x00-\x7f]', ' ', cleanTxt)

  cleanTxt = re.sub('\s+', ' ', cleanTxt)

  return cleanTxt
```

```python
cleanResume('my #### $ # #abhishek website like is this http://heloword and access it @gmail.com')
```

```python
data['Resume'] = data['Resume'].apply(lambda x: cleanResume(x))
```

```python
data['Resume'][0]
```

# Words into categorical **values**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(data['Category'])
```

```python
data['Category'] = le.transform(data['Category'])
```

```python
data['Category'].unique()
```

# **Vecrtorization**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(data['Resume'])
requiredText = tfidf.transform(data['Resume'])
```

```python
requiredText
```

# **Splitting**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(requiredText, data['Category'], random_state=42, test_size=0.2)
```

# Training the **model**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
```

# predicting the **model**

```python
ypred = clf.predict(X_test)
print(ypred)
```

```python
print(accuracy_score(y_test, ypred))
```

# save Files

```python
import pickle
pickle.dump(tfidf, open('tfidf.pkl', 'wb'))
pickle.dump(clf, open('clf.pkl', 'wb'))
```
